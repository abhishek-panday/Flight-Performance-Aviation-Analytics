# Now we are changing the format like 2359 into 23:59 

```python
import pandas as pd
from datetime import datetime, time

# Load the CSV file
flights = pd.read_csv("flightsFinal.csv")

# Function to convert numeric time values into proper HH:MM time format
def convert_to_time(value):
    try:
        value = int(float(value))  # Handle float and ensure it's an integer
        value_str = str(value).zfill(4)  # Ensure 4-digit format (e.g., "5" → "0005")
        hours = int(value_str[:2])  # First two digits are hours
        minutes = int(value_str[2:])  # Last two digits are minutes
        return time(hours, minutes)  # Return as datetime.time type
    except:
        return time(0, 0)  # Default to "00:00" in case of errors

# List of columns to convert
columns_to_convert = ["SCHEDULED_DEPARTURE", "DEPARTURE_TIME" , "WHEELS_ON", "WHEELS_OFF", "ARRIVAL_TIME" , "SCHEDULED_ARRIVAL"]

# Convert each column to proper time format
for col in columns_to_convert:
    if col in flights.columns:
          flights[col] = flights[col].apply(convert_to_time)

```
