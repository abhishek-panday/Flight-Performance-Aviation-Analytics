```python
flights['TAIL_NUMBER'].fillna('UNKNOWN', inplace=True)
```

# All the null values in "TAIL_NUMBER" has been replaced by "Unknown" as there no additional information for Tail number.

```python
# Fill null values for cancelled flights with 0
flights.loc[flights["CANCELLED"] == 1, ["DEPARTURE_TIME", "DEPARTURE_DELAY"]] = 0

# Compute median values for non-cancelled flights
mean_departure_time = flights.loc[flights["CANCELLED"] == 0, "DEPARTURE_TIME"].mean()
mean_departure_delay = flights.loc[flights["CANCELLED"] == 0, "DEPARTURE_DELAY"].mean()

# Fill null values for non-cancelled flights with mean values
flights.loc[flights["CANCELLED"] == 0, "DEPARTURE_TIME"] = flights.loc[flights["CANCELLED"] == 0, "DEPARTURE_TIME"].fillna(mean_departure_time)
flights.loc[flights["CANCELLED"] == 0, "DEPARTURE_DELAY"] = flights.loc[flights["CANCELLED"] == 0, "DEPARTURE_DELAY"].fillna(mean_departure_delay)
```

```python
flights.isnull().sum()
```

# All the null values in "DEPARTURE_TIME" and "DEPARTURE_DELAY" which were null due to cancellation of flights has been replaced by a placeholder that is 0, and the rest null values which were related to cancellation of flights were fileed with the mean of "DEPARTURE_TIME" and "DEPARTURE_DELAY" as there were no outliers in the dataset.

```python
# Fill null values for cancelled flights with 0
flights.loc[flights["CANCELLED"] == 1, ["TAXI_OUT", "WHEELS_OFF"]] = 0

# Compute median values for non-cancelled flights
mean_taxi_out = flights.loc[flights["CANCELLED"] == 0, "TAXI_OUT"].mean()
mean_wheels_off = flights.loc[flights["CANCELLED"] == 0, "WHEELS_OFF"].mean()

# Fill null values for non-cancelled flights with median values
flights.loc[flights["CANCELLED"] == 0, "TAXI_OUT"] = flights.loc[flights["CANCELLED"] == 0, "TAXI_OUT"].fillna(mean_taxi_out)
flights.loc[flights["CANCELLED"] == 0, "WHEELS_OFF"] = flights.loc[flights["CANCELLED"] == 0, "WHEELS_OFF"].fillna(mean_wheels_off)
```

```python
flights.isnull().sum()
```

# All the null values in "TAXI_OUT" and "WHEELS_OFF" which were null due to cancellation of flights has been replaced by a placeholder that is 0, and the rest null values which were related to cancellation of flights were filLed with the mean of "TAXI_OUT" and "WHEELS_OFF" as there were no outliers in the dataset.

```python
# Fill null values for cancelled flights with 0
flights.loc[flights["CANCELLED"] == 1, ["ELAPSED_TIME", "AIR_TIME"]] = 0

# Compute median values for non-cancelled flights
mean_elapsed_time = flights.loc[flights["CANCELLED"] == 0, "ELAPSED_TIME"].mean()
mean_air_time = flights.loc[flights["CANCELLED"] == 0, "AIR_TIME"].mean()

# Fill missing ELAPSED_TIME and AIR_TIME using median
flights.loc[flights["CANCELLED"] == 0, "ELAPSED_TIME"] = flights.loc[flights["CANCELLED"] == 0, "ELAPSED_TIME"].fillna(mean_elapsed_time)
flights.loc[flights["CANCELLED"] == 0, "AIR_TIME"] = flights.loc[flights["CANCELLED"] == 0, "AIR_TIME"].fillna(mean_air_time)
```

```python
flights.isnull().sum()
```

# All the null values in "ELAPSED_TIME" and "AIR_TIME" which were null due to cancellation of flights has been replaced by a placeholder that is 0, and the rest null values which were related to cancellation of flights were fileed with the mean of "ELAPSED_TIME" and "AIR_TIME" as there were no outliers in the dataset.

```python
# Fill null values for cancelled flights with 0
flights.loc[flights["CANCELLED"] == 1, ["WHEELS_ON", "TAXI_IN"]] = 0

# Compute median values for non-cancelled flights
mean_wheels_on = flights.loc[flights["CANCELLED"] == 0, "WHEELS_ON"].mean()
mean_taxi_in = flights.loc[flights["CANCELLED"] == 0, "TAXI_IN"].mean()

# Fill missing WHEELS_ON and TAXI_IN using median
flights.loc[flights["CANCELLED"] == 0, "WHEELS_ON"] = flights.loc[flights["CANCELLED"] == 0, "WHEELS_ON"].fillna(mean_wheels_on)
flights.loc[flights["CANCELLED"] == 0, "TAXI_IN"] = flights.loc[flights["CANCELLED"] == 0, "TAXI_IN"].fillna(mean_taxi_in)
```

```python
flights.isnull().sum()
```

# All the null values in "WHEELS_ON" and "TAXI_IN" which were null due to cancellation of flights has been replaced by a placeholder that is 0, and the rest null values which were related to cancellation of flights were fileed with the mean of "WHEELS_ON" and "TAXI_IN" as there were no outliers in the dataset.

```python
# Fill null values for cancelled flights with 0
flights.loc[flights["CANCELLED"] == 1, ["ARRIVAL_TIME", "ARRIVAL_DELAY"]] = 0

# Compute median values for non-cancelled flights
mean_arrival_time = flights.loc[flights["CANCELLED"] == 0, "ARRIVAL_TIME"].mean()
mean_arrival_delay = flights.loc[flights["CANCELLED"] == 0, "ARRIVAL_DELAY"].mean()

# Fill missing ARRIVAL_TIME and ARRIVAL_DELAY using median
flights.loc[flights["CANCELLED"] == 0, "ARRIVAL_TIME"] = flights.loc[flights["CANCELLED"] == 0, "ARRIVAL_TIME"].fillna(mean_arrival_time)
flights.loc[flights["CANCELLED"] == 0, "ARRIVAL_DELAY"] = flights.loc[flights["CANCELLED"] == 0, "ARRIVAL_DELAY"].fillna(mean_arrival_delay)
```

```python
# Fill missing values in CANCELLATION_REASON for non-cancelled flights
flights.loc[flights["CANCELLED"] == flights["CANCELLATION_REASON"].isnull(), "CANCELLATION_REASON"] = "Not Cancelled"
```

```python
flights.isnull().sum()
```

# The null values in "CANCELLATION_REASON" likely correspond to non-cancelled flights.
# Since "CANCELLATION_REASON" is only relevant for cancelled flights, we filled missing values for non-cancelled flights with "Not Cancelled".

```python
# Fill null values with 0
delay_columns = ['AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY']
flights[delay_columns] = flights[delay_columns].fillna(0)
```

```python
flights.isnull().sum()
```

If there is no delay, these columns are null.
If there is a delay, the corresponding column will have the duration of the delay (in minutes), and the other columns will have 0 (indicating no contribution to the delay from those factors).

Null Values: Indicate no delay.
0 Values: Indicate no contribution to the delay from that specific factor.
Positive Values: Indicate the duration of the delay caused by that specific factor.

# Fill Nulls with 0
# Replaced null values with 0 to indicate no delay or no contribution to the delay.
# This is the most straightforward approach and ensures consistency in the dataset.

```python
# Calculate mean SCHEDULED_TIME for each route
route_means = flights.groupby(['ORIGIN_AIRPORT', 'DESTINATION_AIRPORT'])['SCHEDULED_TIME'].transform('mean')

# Fill null SCHEDULED_TIME values with route-specific means
flights['SCHEDULED_TIME'] = flights['SCHEDULED_TIME'].fillna(route_means)
```

```python
flights['CANCELLATION_REASON'].fillna('UNKNOWN', inplace=True)
```

```python
flights.isnull().sum()
```

We shall impute with Mean of Similar Flights
Calculated the mean SCHEDULED_TIME for flights with the same route (same ORIGIN_AIRPORT and DESTINATION_AIRPORT) and use it to fill null values.

# This approach preserves data and ensures imputed values are contextually relevant.

```python
flights['DIVERTED'] = flights['DIVERTED'].astype(bool)
flights['CANCELLED'] = flights['CANCELLED'].astype(bool)
```
