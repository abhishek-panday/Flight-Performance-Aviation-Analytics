```python
# Calculate mean LATITUDE and LONGITUDE for all airports
mean_latitude = data['LATITUDE'].mean()
mean_longitude = data['LONGITUDE'].mean()

# Fill null values with the mean
data['LATITUDE'] = data['LATITUDE'].fillna(mean_latitude)
data['LONGITUDE'] = data['LONGITUDE'].fillna(mean_longitude)
```
